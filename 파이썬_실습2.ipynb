# 넘파이(umpy)

```python
# 필요한 라이브러리 임포트
import numpy as np
```

## 넘파이 배열과 파이썬 리스트 비교

```python
# 파이썬 리스트 자료 구조 데이터 생성
list_data = [1,2,3]
print(f'파이썬 리스트: {list_data}')

print('-'*80)

# 넘파이 배열 생성
array_data = np.array(list_data)
print(f'넘파이 배열로 변한 후 데이터: {array_data}')
```

```python
# 파이썬 리스트와 넘파이 배열 차이점 확인

## 덧셈 연산

### 파이썬 리스트
result1 = list_data + list_data
print(f'파이썬 리스트에 대한 덧셈 연산의 결과: \n{result1}')

print('-'*80)

### 넘파이 배열
result2 = array_data + array_data
print(f'넘파이 배열에 대한 덧셈 연산의 결과: \n{result2}')

print('-'*80)

## 곱셈 연산

### 파이썬 리스트
result3 = []

for num in list_data:
    result3.append(num*3)

print(f'파이썬 리스트에 대한 곱셈 연산의 결과: \n{result3}')

print('-'*80)

### 넘파이 배열
result4 = array_data * 3
print(f'넘파이 배열에 대한 곱셈 연산의 결과: \n{result4}')

print('-'*80)

## 통계 처리(평균 값 구하기)

### 파이썬 리스트
list_average = sum(list_data) / len(list_data)
print(f'파이썬 리스트에 대한 평균 값: {list_average}')

print('-'*80)

### 넘파이 배열
array_average = array_data.mean()
print(f'넘파이 배열에 대한 평균 값: {array_average}')
```

```python
# 대괄호가 중첩된 리스트 자료 구조 데이터 생성(조건: 성분 원소의 개수 동일)
list_data = [[1,2,3],[4,5,6],[7,8,9]]
print(f'대괄호가 중첩된 리스트 자료 구조 데이터 확인: \n{list_data}')

print('-'*80)

# 넘파이 배열로 변환 -> 엑셀과 같은 표(table: 행과 열) 구조 생성
array_data = np.array(list_data)
print(f'대괄호가 중첩된 리스트를 넘파이 배열로 변환한 결과: \n{array_data}')
```

## 2차원 배열 생성

```python
# 대괄호가 중첩된 리스트 자료 구조 데이터 생성(조건: 성분 원소의 개수 동일)
list_data = [[1,2,3],[4,5,6]]
print(f'대괄호가 중첩된 리스트 자료 구조 데이터 확인: \n{list_data}')

print('-'*80)

# 넘파이 배열로 변환 -> 2차원 배열 -> 엑셀과 같은 표(table: 행과 열) 구조 생성
array_data = np.array(list_data)
print(f'대괄호가 중첩된 리스트를 넘파이 배열로 변환한 결과: \n{array_data}')
```

## 인덱싱

```python
# 목표: 첫 번째 행, 첫 번째 열 -> 1 추출

## 파이썬 방식

### 1단계: 변수[인덱스] -> 첫 번째 행 추출
result1 = array_data[0]
print(result1)

print('-'*80)

### 2단계: 변수[인덱스] -> 첫 번째 성분(첫 번째 열) 원소
# result2 = result1[0]
result2 = array_data[0][0]
print(result2)

print('-'*80)

## 넘파이 방식: 넘파이의 배열 모양(2행,3열) 이용 -> 변수[행 인덱스, 열 인덱스]
result3 = array_data[0,0]
print(f'숫자 1을 인덱싱한 결과: {result3}')
```
